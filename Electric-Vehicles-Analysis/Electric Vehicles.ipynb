## Drop the Unnessessary Columns

```python
# VIN (1-10) , Legislative District , DOL Vehicle ID

data.drop(columns = ['VIN (1-10)' , 'Legislative District' , 'DOL Vehicle ID', 'Electric Utility.1'] , inplace = True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
columns_with_null = ['County', 'City', 'Postal Code', 'Vehicle Location', 'Electric Utility']
for i in columns_with_null:
    if columns_with_null == 'Postal Code':
        data[i].fillna(0, inplace = True)
    else:
        data[i].fillna('Unknown', inplace = True)
```

```python
data['Make'].value_counts()
```

```python
top_10_makes = data['Make'].value_counts().nlargest(20).index

sns.countplot(data=data, x='Make', order=top_10_makes, palette='viridis')
plt.xticks(rotation=90)
plt.title('Count of Vehicles by Make')
plt.show()
```

```python
labels = data['Make'].value_counts().nlargest(10).index
sizes = data['Make'].value_counts().nlargest(10).values
colors = ["#931515","#164675","#086508","#764412","pink","yellow"]
plt.figure(figsize = (8,8), facecolor='none')
plt.pie(sizes, labels=labels, rotatelabels=False, shadow=True, colors=colors, explode=[0.1,0.03,0.03,0.03,0.03,0.03,0.03,0.03,0.03,0.03], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Vehicle Makes Distribution', color='blue', fontsize=20)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

text = ' '.join(data['Make'])

plt.rcParams['figure.figsize'] = (12,12)
wordcloud = WordCloud(background_color = 'black',colormap='vlag', width = 1200,  height = 1200, max_words = 121).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

print(text)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'data' is your DataFrame with columns: 'Make', 'Model', 'Electric Range'

# Get top 20 makes
top_20_makes = data['Make'].value_counts().nlargest(20).index

# Filter data for top 20 makes
filtered_data = data[data['Make'].isin(top_20_makes)]

# Get top 10 models within these makes
top_10_models = filtered_data['Model'].value_counts().nlargest(10).index

# Filter data for top 10 models
filtered_data = filtered_data[filtered_data['Model'].isin(top_10_models)]

# Create the plot
plt.figure(figsize=(20, 20))
sns.boxplot(data=filtered_data, x="Model", y="Electric Range", palette="twilight")

plt.xticks(rotation=90)
plt.title('Electric Range Distribution by Model (Top 20 Makes)')
plt.xlabel('Model')
plt.ylabel('Electric Range (miles)')
plt.tight_layout()
plt.show()
```

<p style="padding: 10px;

```python
df1 = data.groupby(['Electric Vehicle Type', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Make'], as_index=False)[['Electric Range']].mean()

fig = px.treemap(df1, path=[px.Constant("all"), 'Electric Vehicle Type', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Make'], values='Electric Range',
                 color='Electric Range',
                 color_continuous_scale='RdBu')

fig.update_layout(margin  = dict(t=10, l=10, r=10, b=10))
fig.show()
```
