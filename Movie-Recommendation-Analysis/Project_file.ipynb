## IMDB Weighted Rating Formula

$$W = \frac{Rv + Cm}{v + m}$$

**Where:**

- $W$ = Weighted Rating 
- $R$ = average for the movie as a number from 0 to 10 (mean) = (Rating)
- $v$ = number of votes for the movie = (votes)  
- $m$ = minimum votes required to be listed in the Top 250 (currently 3000)
- $C$ = the mean vote across the whole report (currently 6.9)

```python
movies_cleaned_df.columns
```

```python
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
```

```python
movies_cleaned_df[['original_title', 'vote_count']]
```

```python
px.box(movies_cleaned_df, x='vote_count', title='Box Plot of Vote Count')
```

m = 1900

```python
threshold = movies_cleaned_df['vote_count'].quantile(q=0.9)
threshold
```

```python
movies_cleaned_df['vote_count'] >= threshold
```

```python
filtered_df = movies_cleaned_df[movies_cleaned_df['vote_count'] >= threshold]
filtered_df.head(3)
```

```python
# R, v, C, m
```

```python
m = threshold
m
```

```python
R
```

```python
C
```

```python
v
```

```python
v = filtered_df['vote_count']
R = filtered_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
```

```python
v
```

```python
C
```

```python
R
```

```python
filtered_df['weighted_avg'] = ((R*v)+(C*m))/(v+m)
filtered_df.head(3)
```

Top 20 most recommended movies based on average weighted :

```python
df_sorted_ranking = filtered_df.sort_values('weighted_avg', ascending=False).head(20)
df_sorted_ranking.head(3)
```

Build Hybrid Recommendation Feature <br>
<span style="font-size:50%;">(based on weighted average & Popularity score)</span>
</p>

```python
filtered_df.columns
```

```python
filtered_df['score_mix'] = filtered_df['weighted_avg_scaled']*0.5 + filtered_df['popularity_scaled']*0.5
```

```python
filtered_df.head(2)
```

```python
recommended_movies = filtered_df.sort_values('score_mix', ascending=False)[0:20]
recommended_movies.head(3)
```

```python
print(recommended_movies[['score_mix']].head())
```

```python
fig_mix = px.bar(
            recommended_movies,
            x='score_mix',
            y='original_title',
            orientation='h',
            color='score_mix',
            color_continuous_scale=['#bda0ca', '#4a006c']
            )

fig_mix.update_layout(yaxis={'categoryorder':'total ascending'})

fig_mix.show()
```

#### 🎬 Insight:
- Interstellar tops the combined score_mix (~0.88), balancing both high rating and popularity
- Minions and Guardians of the Galaxy follow closely, showing that mainstream blockbusters
- score well when both popularity and rating are combined

##### This plot shows the weighted average, popularity, and mix score for each movie :

```python
import plotly.express as px

fig = px.bar(
    recommended_movies,
    y='original_title',
    x='score_mix',
    orientation='h',
    color='score_mix',
    color_continuous_scale=['#bda0ca', '#4a006c'],

    hover_data={
        'weighted_avg': True,
        'popularity': True,
        'score_mix': True,
        'original_title': False
    }
)

fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    # xaxis_tickangle=65,
    width=900,
    height=600
)

fig.show()
```

#### 🎬 Insight:

- Scores drop gradually from ~0.65 to 0.50, suggesting most recommended movies are closely matched

#### - Interstellar tops the recommendation list with the highest score_mix (approximately 0.88). This confirms it perfectly balances a very high rating with strong popularity.
#### - Minions and Guardians of the Galaxy come in second and third, showing that light, entertaining blockbuster movies perform exceptionally well when popularity is combined with ratings.
#### - The Dark Knight and The Shawshank Redemption remain among the highest-rated films even years after their release, proving the lasting power of classic movies in the recommendation system.
#### - Movies that achieve a strong balance between high ratings and high popularity (such as Inception, Fight Club, and Mad Max: Fury Road) consistently rank at the top when using Weighted Average + Popularity together.
#### - The score_mix varies significantly across movies, starting from 0.88 for the best film and gradually dropping to around 0.50. This indicates that the recommendations are diverse and not limited to just one genre or type.
#### - High-rated and highly popular films like Star Wars and The Lord of the Rings receive high priority in recommendations, making the system much smarter than relying on rating alone.
#### - The system proves effective at recommending a wide variety of movies (Drama, Action, Animation, Sci-Fi) as long as they are strong in both rating and popularity.
